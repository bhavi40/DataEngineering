```python
import requests
import json

# GitHub repo details
username = "bhavi40"
repo = "DataEngineering"
folder_path = "Brazillian%20ecommerce/Data"

# GitHub API URL for the folder
api_url = f"https://api.github.com/repos/{username}/{repo}/contents/{folder_path}"

# Get folder content
response = requests.get(api_url)
data = response.json()

# Create the list in the required format
result = [
    {
        "csv_relative_url": f"{repo}/refs/heads/main/{folder_path}/{file['name']}",
        "file_name": file['name']
    }
    for file in data if file['name'].endswith('.csv')
]

# Pretty print the JSON
print(json.dumps(result, indent=4))
```
