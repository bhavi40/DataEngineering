```python
from pymongo import MongoClient
from urllib.parse import quote_plus
import pandas as pd

username = "mongodbadmin"
password = quote_plus("Team@123")  # encodes @ correctly

# ✅ Use the global URI directly
uri = f"mongodb+srv://{username}:{password}@ecomm-mongodb.global.mongocluster.cosmos.azure.com/?tls=true&authMechanism=SCRAM-SHA-256&retrywrites=false&maxIdleTimeMS=120000"

# Create client
client = MongoClient(uri)

# Choose database and collection
db = client["OlistDataNoSQL"]
collection = db["olist_order_payments_dataset"]

# Load CSV and insert
df = pd.read_csv("product_category_name_translation.csv")
data = df.to_dict(orient="records")

batch_size = 500
for i in range(0, len(data), batch_size):
    collection.insert_many(data[i:i + batch_size])

print("✅ CSV uploaded successfully to Cosmos DB MongoDB!")

```
